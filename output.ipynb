```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data = pd.read_csv('./output4.csv')
```

```python
def cls(n):
    mask, gender, age='','',''
    if n<6: mask='Wear'
    elif n<12: mask='Incorrect'
    else: mask='Not Wear'
    
    if (n//3)%2==0: gender='Male'
    else: gender='Female'
    
    if n%3==0: age='<30'
    elif n%3==1: age='30~59'
    else: age='60>='
    
    return mask, gender, age
    
```

```python
plt.figure(figsize=(20,30))
for i, idx in enumerate(np.random.choice(12600,64)):
    plt.subplot(8, 8, i+1)
    img = Image.open('./input/data/eval/images/' + data.iloc[idx][0])
    plt.imshow(np.asarray(img))
    plt.axis('off')
    t=cls(data.iloc[idx][1])
    plt.title('\n'.join(t))
plt.show()    
print ("Done")
```
